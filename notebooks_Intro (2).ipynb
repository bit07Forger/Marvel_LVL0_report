## Quest 2: Python Coding & Visualization
### Introduction
In this section, I will declare variables, perform calculations, and plot a graph using Matplotlib.

````python
# Declaring two variables
length = 15
width = 8

# Calculating area
area = length * width
print(f"Length: {length}")
print(f"Width: {width}")
print(f"Area of Rectangle: {area}")
````

````python
import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [200, 350, 300, 480, 420]

# Plot
plt.plot(months, sales, marker='o', color='steelblue', linewidth=2)
plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()




````
